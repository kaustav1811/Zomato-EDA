```python
#Analyzing price range
plt.figure(figsize=(10, 6))
sns.histplot(final_df['Price range'], bins=5, kde=True)
plt.title('Distribution of Price Range')
plt.xlabel('Price Range')
plt.ylabel('Frequency')
plt.show()
```

```python
#Exploring relationship between rating and price range
plt.figure(figsize=(10, 6))
sns.boxplot(x='Price range', y='Aggregate rating', data=final_df)
plt.title('Rating vs. Price Range')
plt.show()
```

```python
#Analyzing Average Cost for Two across Cities/Countrie
avg_cost_by_city = final_df.groupby('City')['Average Cost for two'].mean().sort_values(ascending=False)
print(avg_cost_by_city.head(10))
avg_cost_by_country = final_df.groupby('Country')['Average Cost for two'].mean().sort_values(ascending=False)
print(avg_cost_by_country.head(10))
```

```python
#Analyzing a specific cuisine
north_indian_restaurants = final_df[final_df['Cuisines'].str.contains('North Indian', na=False)]
plt.figure(figsize=(10, 6))
sns.histplot(north_indian_restaurants['Price range'], bins=5, kde=True)
plt.title('Distribution of Price Range for North Indian Restaurants')
plt.xlabel('Price Range')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(final_df[['Aggregate rating', 'Votes', 'Average Cost for two', 'Price range']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.regplot(x='Average Cost for two', y='Aggregate rating', data=final_df)
plt.title('Relationship between Average Cost and Aggregate Rating (with Regression Line)')
plt.xlabel('Average Cost for Two')
plt.ylabel('Aggregate Rating')
plt.show()
```

```python
sns.regplot(x='Average Cost for two', y='Aggregate rating', data=final_df)
plt.xscale('log')  #using log scale to avoid clutter by outliers

plt.title('Relationship between Average Cost and Aggregate Rating (Log Scale)')
plt.xlabel('Average Cost for Two (Log Scale)')
plt.ylabel('Aggregate Rating')
plt.show()
```

```python
# Average Cost by City (Top 10)
avg_cost_by_city = final_df.groupby('City')['Average Cost for two'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=avg_cost_by_city.index, y=avg_cost_by_city.values)
plt.title('Average Cost for Two by City (Top 10)')
plt.xlabel('City')
plt.ylabel('Average Cost for Two')
plt.xticks(rotation=30, ha='right')
plt.show()
```

```python
# Average Cost by Country (Top 10)
avg_cost_by_country = final_df.groupby('Country')['Average Cost for two'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=avg_cost_by_country.index, y=avg_cost_by_country.values)
plt.title('Average Cost for Two by Country (Top 10)')
plt.xlabel('Country')
plt.ylabel('Average Cost for Two')
plt.xticks(rotation=30, ha='right')
plt.yscale('log') #To overcome outlier
plt.show()
```
